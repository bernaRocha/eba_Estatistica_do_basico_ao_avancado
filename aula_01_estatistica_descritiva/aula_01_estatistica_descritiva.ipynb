```python
media = df.SalePrice.mean()
mediana = df.SalePrice.median()
moda = df.SalePrice.mode()
```

```python
df.SalePrice.describe()
```

```python
df['categoria_preco'] = np.where(df.SalePrice > 214000, 'cara', 'barata')
df
```

```python
fig = sns.boxplot(x='categoria_preco', y='OverallQual', data=df)

print(f"Média: {df.groupby('categoria_preco')['OverallQual'].mean()}")
print(f"Mediana: {df.groupby('categoria_preco')['OverallQual'].median()}")

plt.xticks(rotation=90);
```

```python
fig = sns.boxplot(x='categoria_preco', y='OverallCond', data=df)

print(f"Média: {df.groupby('categoria_preco')['OverallCond'].mean()}")
print(f"Mediana: {df.groupby('categoria_preco')['OverallCond'].median()}")

plt.xticks(rotation=90);
```

```python
# Verificando a categoria cara
df[df.categoria_preco == "cara"][['categoria_preco', 'OverallCond']]
```

```python
df[df.categoria_preco == "cara"].OverallCond.value_counts()
```

```python
df[df.categoria_preco == "cara"].OverallCond.value_counts(normalize=True) * 100
```

```python
fig = sns.boxplot(x='categoria_preco', y='TotRmsAbvGrd', data=df)
print(f'Média: {df.groupby("categoria_preco")["TotRmsAbvGrd"].mean()}')
print(f'Mediana: {df.groupby("categoria_preco")["TotRmsAbvGrd"].median()}')
plt.xticks(rotation=90);
```

```python
fig = sns.boxplot(x="categoria_preco", y="YearBuilt", data=df)
print("media:", df.groupby("categoria_preco")["YearBuilt"].mean())
print("mediana:", df.groupby("categoria_preco")["YearBuilt"].median())
plt.xticks(rotation=90);
```

## Skew

Calculando o skew conseguimos entender onde os outliers estão mais concentrados

```python
#assimetria
from scipy.stats import skew
skew(df['SalePrice'])
```

Vemos aqui uma assimetria positiva, indicando que os outliers estão concentrados a direita. De fato, comprovamos isso vendo o histograma
